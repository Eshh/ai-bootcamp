```python
names = ['K', 'J', 'I', 'D', 'E']
ages = [25, 30, 35, 28, 22]
salaries = [50000, 60000, 75000, 55000, 45000]
departments = ['IT', 'HR', 'IT', 'Finance', 'IT']
```

```python
def age_filter(num):
  if num < 30:
    return 'young'
  return 'senior'
df['age_group_new'] = df['age'].apply(age_filter)
df.head()
```

```python
def required_filter(row):
  age = row['age']
  department = row['department']
  if age <= 30 and department == 'IT':
    return 'IT Junior'
  return 'senior'

df['staff_seg'] = df.apply(required_filter, axis=1)
df.head()
```
